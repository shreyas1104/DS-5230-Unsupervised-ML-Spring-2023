```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_openml
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.manifold import TSNE
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report
import random
```

```python
def run_tSNE(X, n, p):

    tsne = TSNE(n_components=n, perplexity=p)
    X_tsne = tsne.fit_transform(X)

    return tsne, X_tsne
```

### MNIST

```python
mnist = fetch_openml('mnist_784')
```

```python
X = mnist.data
label = mnist.target
df = pd.concat([X, label], axis=1).rename(columns={'class':'label'})
mnist_df = df.groupby('label', group_keys=False).apply(pd.DataFrame.sample, n=1000)
mnist_df = mnist_df.sample(frac=1)

X = mnist_df.iloc[:, :-1]
label = mnist_df.iloc[:, -1]
```

#### Perplexity = 5

```python
tsne, X_tsne = run_tSNE(X, n=2, p=5)
```

```python
tsne_result_df = pd.DataFrame({'x': X_tsne[:,0], 'y': X_tsne[:,1], 'label': label})
plt.figure(figsize=(10,10))
sns.scatterplot(x='x', y='y', hue='label', data=tsne_result_df, legend='full')
```

```python
tsne, X_tsne = run_tSNE(X, n=3, p=5)
```

```python
tsne_result_df = pd.DataFrame({'x': X_tsne[:,0], 'y': X_tsne[:,1], 'z': X_tsne[:,2], 'label': label.astype(int)})
fig = plt.figure(figsize=(15,12))
ax = fig.add_subplot(projection = '3d')
ax.scatter3D(tsne_result_df.x, tsne_result_df.y, tsne_result_df.z, c=tsne_result_df.label)
ax.set_xlabel ('x')
ax.set_ylabel ('y')
ax.set_zlabel ('z')
plt.show()
```
