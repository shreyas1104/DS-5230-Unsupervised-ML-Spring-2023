```python
import pandas as pd
import numpy as np
from google.colab import drive
import random
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
data = pd.read_csv('/content/drive/MyDrive/UML/HW6/ml-100k/u.data', header= None , sep = '\t')
data.columns = ['user_id' , 'movie_id' , 'rating' , 'timestamp']
```

```python
data.head()
```

```python
user_movie = data[['user_id', 'movie_id', 'rating']].pivot(index='user_id', columns='movie_id', values='rating').fillna(0).reset_index(drop=False).rename_axis(None, axis=1)
```

```python
user_movie.head()
```

```python
user_user_sim = pd.DataFrame(cosine_similarity(user_movie, user_movie))
```

```python
user_user_sim.head()
```

```python
len_train = 800
len_test = len(user_movie) - len_train
```

```python
train_idx = list(np.arange(1, (len_train+1)))
test_idx = list(np.arange((len_train+1), (len_train+len_test+1)))
```

```python
train_df = user_movie[user_movie['user_id'].isin(train_idx)]
test_df = user_movie[user_movie['user_id'].isin(test_idx)]
```

```python
train_df
```

```python
test_df
```

```python
test_sim = user_user_sim.iloc[len_train:, :len_train]
test_sim
```

```python
total_rmse = 0

for i in range(len_test):

    actual_ratings = test_df.iloc[i, 1:].tolist()
    predicted_ratings = []

    for mid in range(1, 1683):
        sim_weights = test_sim.iloc[i].tolist()
        train_ratings = train_df.iloc[:, mid].tolist()
        pred = np.average(train_ratings, weights=sim_weights)
        predicted_ratings.append(pred)

    total_rmse += sqrt(mean_squared_error(actual_ratings, predicted_ratings))

avg_rmse = total_rmse / len_test
```
